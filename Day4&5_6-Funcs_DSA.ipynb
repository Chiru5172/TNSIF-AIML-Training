```python
def arm(n):
  t=0
  temp=n
  l=len(str(n))
  while temp>0:
    t+=(temp%10)**l
    temp//=10
  if t==n:
    print("Armstrong number")
  else:
    print("Not an armstrong number")
n=int(input())
arm(n)
```

```python
def fib(n):
  a,b=0,1
  for i in range(n):
    print(a)
    a,b=b,a+b
fib(20)
```

```python
def add(n):
  s=0
  while n>0:
    rem=n%10
    s+=rem
    n//=10
  print(s)
add(123)
```

```python
%%writefile area.py
import math
def circle_Area(radius):
  return 3.14*radius*radius
def rectangle_area(length,breadth):
  return length*breadth
def triangle(base,height):
  return 0.5*base*height
```
