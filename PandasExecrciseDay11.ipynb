```python
import pandas as pd
df=pd.read_csv('/content/students_dataset.csv')
df
```

```python
df[df['marks']>80]
```

```python
df.sort_values(by='marks',ascending=False)
```

```python
grades=[]
for mark in df['marks']:
  if mark>=90:
    grades.append('A')
  elif mark>=80:
    grades.append('B')
  elif mark>=70:
    grades.append('C')
  elif mark>=60:
    grades.append('D')
  else:
    grades.append('F')
df['grades']=grades
df
```

```python
df["name"]=df['name'].str.upper()
df
```

```python
df=df.drop(columns='department')
df
```

```python
df.drop(index=3)
```

```python
df.groupby('department')['marks'].mean()
```

```python
df1=df['marks'].agg(['sum','mean','max','min','median','std'])
print(df1)
```

```python
df2={'name':['Teja','Divya'],'age':[20,21],'marks':[87,91],'department':['CSE','ECE']}
pd.concat([df,pd.DataFrame(df2)])
```

```python
df['exam_date']=pd.to_datetime(df['exam_date'])
df['day']=df['exam_date'].dt.day
df['month']=df['exam_date'].dt.month
df['year']=df['exam_date'].dt.year
print(df[['exam_date','day','month','year']])
```
