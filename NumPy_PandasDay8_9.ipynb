**Pandas**

```python
import pandas as pd
df=pd.DataFrame({"SNo":[1,2,3],"Name":["A","B","C"]})
print(df)
```

```python
import pandas as pd
df=pd.read_csv('/content/students.csv')
print("Top 5 rows")
print(df.head(5))
print("Last 5 rows")
print(df.tail(5))
print("Column Names")
print(df.columns)
print(df.info())
```

```python
marks_info=df[df['marks']>85]
print(marks_info)
```

```python
df1=df[df['city']=='Hyderabad']
print(df1)
```

```python
df2=df[df['gender']=='Male']
print(df2)
```

```python
df3=df.groupby('city')['marks'].mean().sort_values(ascending=False)
print("City wise average marks")
print(df3)
print(df3.head(1))
```
