```python
import pandas as pd
df=pd.read_csv('employees_dataset.csv')
df
```

```python
df.isna().sum()
```

```python
df["name"].fillna("Unknown", inplace=True)
```

```python
df
```

```python
df["department"] = df["department"].fillna(df["department"].mode()[0])
```

```python
df
```

```python
df["salary"] = df["salary"].fillna(df["salary"].median())
```

```python
df
```

```python
df["years_of_experience"] = df["years_of_experience"].fillna(df["years_of_experience"].mean())
```

```python
df
```

```python
df['name']=df['name'].str.lower()
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df[df["department"] == "IT"]
```

```python
df[df['salary']>70000]
```

```python
df[df['years_of_experience']>5]
```

```python
df[(df["department"] == "IT") | (df["department"] == "Finance")]
```

```python
df[(df['salary']>60000) & (df['salary']<80000)]
```

```python
df.groupby("department").size()
```

```python
df.groupby("department")["salary"].sum()
```

```python
df.groupby("department")["salary"].agg(['mean','max','min','sum'])
```
