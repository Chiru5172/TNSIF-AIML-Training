```python
n=int(input())
t=0
temp=n
l=len(str(n))
while temp>0:
  t+=(temp%10)**l
  temp//=10
if t==n:
  print("Armstrong number")
else:
  print("Not an armstrong number")
```

```python
import math
a = int(input())
b = int(input())
m=a
n=b
while b!=0:
  a,b=b,a%b
gcd = math.gcd(a, b)
lcm = (m*n) // gcd
print("LCM of", m, "and", n, "is:", lcm)
```

```python
while True:
  b=15000
  n=int(input())
  if n==1:
    print("Check balance")
    print("Balance:",b)
  elif n==2:
    print("Deposit Money")
    a=int(input("Amount: "))
    if a>0:
      b+=a
      print("Deposited")
      print("Updated Balance:",b)
    else:
      print("Enter valid amount")
  elif n==3:
    print("Withdraw Money")
    c=int(input("Amount"))
    if c<b:
      print("Withdraw successful")
      b=b+a-c
      print("Updated Balance:",b)
    else:
      print("Enter valid amount")
      break
  elif n==4:
    print("Mini Statement")
  else:
    print("Select valid menu")
    print("Transaction failed")
    break
```

```python
rows = int(input("Enter number of rows: "))

for i in range(rows):
    num = 1

    for j in range(rows - i - 1):
        print(" ", end="")

    for j in range(i + 1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1)

    print()
```
